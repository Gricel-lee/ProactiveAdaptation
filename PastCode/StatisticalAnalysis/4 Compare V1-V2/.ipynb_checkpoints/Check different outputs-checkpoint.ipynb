```python
def compare_files(file1_path, file2_path):
    # Read contents of the first file
    with open(file1_path, 'r') as file1:
        file1_contents = file1.read()

    # Read contents of the second file
    with open(file2_path, 'r') as file2:
        file2_contents = file2.read()

    # Compare contents
    if file1_contents == file2_contents:
        print("Files are identical.")
    else:
        print("Files are different.")

# Example usage
file1_path = "outputV1.txt"
file2_path = "output.txt"
compare_files(file1_path, file2_path)
```

```python
def compare_files(file1_path, file2_path):
    # Read contents of the first file
    with open(file1_path, 'r') as file1:
        file1_lines = file1.readlines()

    # Read contents of the second file
    with open(file2_path, 'r') as file2:
        file2_lines = file2.readlines()

    # Compare line by line
    for i, (line1, line2) in enumerate(zip(file1_lines, file2_lines), start=1):
        if line1 != line2:
            print(f"Line {i} is different:")
            print(f"   File 1: {line1.strip()}")
            print(f"   File 2: {line2.strip()}")

# Example usage
file1_path = "outputV1.txt"
file2_path = "output.txt"
compare_files(file1_path, file2_path)
```

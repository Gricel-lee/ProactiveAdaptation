```python
import pandas as pd

file = "Historical logs/day_degrading_grip.csv"

df = pd.read_csv(file)

df
```

```python
import matplotlib.pyplot as plt

for i in range (1,len(df.columns)):
    plt.figure()
    plt.plot(df[df.columns[0]],df[df.columns[i]])

    # Add labels and title
    plt.xlabel(df.columns[0])
    plt.ylabel(df.columns[i])
    plt.title(f'Plot of {df.columns[0]} vs {df.columns[i]}')
    plt.show
```
